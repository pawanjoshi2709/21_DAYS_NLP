Stock preadction using BOW/TFIDF

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Data.csv", encoding="ISO-8859-1")
```

```python
df.head()
```

```python
train =df[df['Date']<='20150101']
test = df[df['Date']>'20150101']
```

```python
test.shape
```

```python
# remove punctuation
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex = True,inplace =True)

#renaming the columns for easy access

list1=[i for i in range(25)]
new_Index =[str(i) for i in list1]
data.columns = new_Index
data.head()
```

```python
# remove punctuation
data1 = test.iloc[:,2:27]
data1.replace("[^a-zA-Z]"," ",regex = True,inplace =True)

#renaming the columns for easy access

list2=[i for i in range(25)]
new_Index =[str(i) for i in list2]
data1.columns = new_Index
data1.head()
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
  headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[1]
```

```python
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500, ngram_range =(2,2))
traindataset = cv.fit_transform(headlines)
```

```python
test_transform = []
for row in range(0,len(data1.index)):
  test_transform.append(' '.join(str(x) for x in data1.iloc[row,0:25]))
test_dataset = cv.fit_transform(test_transform)
```

```python
# implement RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
# traindataset - X
# train['Label'] - y
```

```python
predictions = randomclassifier.predict(test_dataset)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
